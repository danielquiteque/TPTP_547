```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Parâmetro da distribuição de Poisson (média semanal)
lambda_semanal = 3

# Valores da variável (número de falhas)
x = np.arange(0, 11)
pmf = poisson.pmf(x, mu=lambda_semanal)

# Criar histograma
plt.figure(figsize=(10, 6))
bars = plt.bar(x, pmf, color='lightblue', edgecolor='black')

# Destacar as barras para X >= 3
for i in range(len(x)):
    if x[i] >= 3:
        bars[i].set_color('orange')

# Adiciona os valores no topo das barras
for i, p in enumerate(pmf):
    plt.text(x[i], p + 0.005, f"{p:.2f}", ha='center')

# Configurações do gráfico
plt.title('Distribuição de Poisson (λ = 3)\nNúmero de Falhas por Semana')
plt.xlabel('Número de falhas')
plt.ylabel('Probabilidade P(X = k)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(x)
plt.tight_layout()
plt.show()
```
