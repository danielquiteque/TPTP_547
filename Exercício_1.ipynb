```python
import matplotlib.pyplot as plt

# Parâmetros do GLC
a = 5
c = 3
m = 16
x = 7

# Gerar a sequência completa até repetir
valores = [x]
while True:
    x = (a * x + c) % m
    if x == valores[0]:  # Se voltou ao início, encerra
        break
    valores.append(x)

# Plotar o gráfico
plt.figure(figsize=(10, 5))
plt.plot(range(len(valores)), valores, marker='o')
plt.title("Sequência Gerada pelo GLC Misto")
plt.xlabel("n")
plt.ylabel("X_n")
plt.grid(True)
plt.xticks(range(len(valores)))
plt.yticks(range(m))
plt.show()
```
