```python
import numpy as np
import matplotlib.pyplot as plt

# Número de amostras que desejamos gerar da distribuição f(x) = 3x²
N = 10000

# Constante c calculada anteriormente (máximo de f(x)/g(x))
c = 3

# Define a função densidade de probabilidade alvo f(x) = 3x²
f = lambda x: 3 * x**2

# Define a densidade da distribuição candidata g(x), que é uniforme em [0,1]
g = lambda x: 1  # constante, pois é uniforme

# Lista onde vamos armazenar os valores aceitos de X
samples = []

# Laço de geração de amostras usando o método da aceitação-rejeição
while len(samples) < N:
    # Gera um valor x da distribuição candidata (uniforme entre 0 e 1)
    x = np.random.uniform(0, 1)

    # Gera um valor u da distribuição uniforme entre 0 e 1
    u = np.random.uniform(0, 1)

    # Critério de aceitação: se u <= f(x)/(c * g(x)), aceitamos x
    # Como g(x) = 1 e c = 3, temos: f(x)/(c*g(x)) = (3x²)/3 = x²
    if u <= x**2:
        samples.append(x)  # Adiciona x à lista de amostras aceitas

# Agora vamos visualizar os resultados com um histograma e a PDF original

# Cria uma sequência de valores entre 0 e 1 para traçar a PDF analítica
x_vals = np.linspace(0, 1, 1000)

# Calcula os valores da PDF f(x) = 3x² para os pontos x_vals
pdf_vals = f(x_vals)

# Inicia o gráfico
plt.figure(figsize=(8, 5))

# Plota o histograma das amostras geradas, normalizado para virar uma densidade
plt.hist(samples, bins=50, density=True, alpha=0.6, label='Histograma (amostras)')

# Plota a PDF teórica f(x) = 3x² em vermelho
plt.plot(x_vals, pdf_vals, 'r-', label='PDF analítica: $f(x)=3x^2$')

# Personaliza o gráfico
plt.title('Método da Aceitação-Rejeição para f(x) = 3x²')
plt.xlabel('x')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)

# Mostra o gráfico
plt.show()
```
