```python
# Reimportar as bibliotecas após o reset do ambiente
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# ------------------------------
# Parâmetro da distribuição de Poisson
# λ (lambda) representa a média de ocorrências em um intervalo fixo
lambda_ = 3

# ------------------------------
# Gerar os valores de k (número de ocorrências) de 0 a 10
k_values = np.arange(0, 11)

# Calcular as probabilidades usando a função de massa de probabilidade da Poisson
probabilities = poisson.pmf(k_values, mu=lambda_)

# ------------------------------
# Criar o gráfico de barras (histograma da distribuição)
plt.figure(figsize=(10, 6))
bars = plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')

# Destacar a barra onde k = 5 (exatamente 5 chamadas)
bars[5].set_color('orange')

# Destacar as barras onde k <= 2 (no máximo 2 chamadas)
for i in range(3):  # i = 0, 1, 2
    bars[i].set_color('green')

# ------------------------------
# Adicionar os valores numéricos acima de cada barra
for i, p in enumerate(probabilities):
    plt.text(i, p + 0.005, f"{p:.2f}", ha='center')

# ------------------------------
# Personalizar o gráfico
plt.title('Distribuição de Poisson (λ = 3)\nNúmero de chamadas por minuto')
plt.xlabel('Número de chamadas (k)')
plt.ylabel('Probabilidade P(X = k)')
plt.xticks(k_values)  # Mostrar todos os valores de k no eixo x
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grade no eixo y
plt.tight_layout()
plt.show()
```
