```python
# Reimportar as bibliotecas aps o reset do ambiente
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# ------------------------------
# Parmetro da distribuio de Poisson
#  (lambda) representa a mdia de ocorrncias em um intervalo fixo
lambda_ = 3

# ------------------------------
# Gerar os valores de k (nmero de ocorrncias) de 0 a 10
k_values = np.arange(0, 11)

# Calcular as probabilidades usando a funo de massa de probabilidade da Poisson
probabilities = poisson.pmf(k_values, mu=lambda_)

# ------------------------------
# Criar o grfico de barras (histograma da distribuio)
plt.figure(figsize=(10, 6))
bars = plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')

# Destacar a barra onde k = 5 (exatamente 5 chamadas)
bars[5].set_color('orange')

# Destacar as barras onde k <= 2 (no mximo 2 chamadas)
for i in range(3):  # i = 0, 1, 2
    bars[i].set_color('green')

# ------------------------------
# Adicionar os valores numricos acima de cada barra
for i, p in enumerate(probabilities):
    plt.text(i, p + 0.005, f"{p:.2f}", ha='center')

# ------------------------------
# Personalizar o grfico
plt.title('Distribuio de Poisson ( = 3)\nNmero de chamadas por minuto')
plt.xlabel('Nmero de chamadas (k)')
plt.ylabel('Probabilidade P(X = k)')
plt.xticks(k_values)  # Mostrar todos os valores de k no eixo x
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grade no eixo y
plt.tight_layout()
plt.show()
```
