```python
import numpy as np
import matplotlib.pyplot as plt

# Número de amostras
N = 10000

# Função densidade alvo
f = lambda x: 3 * x**2

# Geração de amostras via método da inversa
u = np.random.uniform(0, 1, N)
samples = u ** (1/3)  # F^-1(u) = u^(1/3)

# Valores para a PDF analítica
x_vals = np.linspace(0, 1, 1000)
pdf_vals = f(x_vals)

# Plot do histograma e da PDF
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=50, density=True, alpha=0.6, label='Histograma (amostras)')
plt.plot(x_vals, pdf_vals, 'r-', label='PDF analítica: $f(x)=3x^2$')
plt.title('Método da Inversa para f(x) = 3x²')
plt.xlabel('x')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
